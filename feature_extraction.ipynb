```python
import nltk

nltk.download('punkt')

def count_sentences(text):
    sentences = nltk.sent_tokenize(text)
    return len(sentences)

df['sent_num'] = df['text'].apply(count_sentences)
```

```python
import nltk
import string

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ –∑–Ω–∞–∫–æ–≤ –ø—Ä–µ–ø–∏–Ω–∞–Ω–∏—è –≤ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–∏
def count_punctuation(sentence):
    count = 0
    for token in nltk.word_tokenize(sentence):
        if token in string.punctuation:
            count += 1
    punctuation_count = count / len(nltk.word_tokenize(sentence))
    return punctuation_count

df['punct_per_sent_num'] = df.text.apply(count_punctuation)
```

```python
import re
```

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –º–Ω–æ–≥–æ—Ç–æ—á–∏–π –≤ —Ç–µ–∫—Å—Ç–µ
def count_ellipsis(text):
    ellipsis_count = len(re.findall('\.{3,}', text))
    return ellipsis_count

# –ü—Ä–∏–º–µ–Ω–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –∫ —Å—Ç–æ–ª–±—Ü—É 'Text' –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ
df['ellipsis_num'] = df['text'].apply(count_ellipsis)
```

```python
def count_capitalized_words(text):
    words = text.split()  # –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ —Ç–µ–∫—Å—Ç–∞ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω—ã–µ —Å–ª–æ–≤–∞
    capitalized_count = 0  # –°—á–µ—Ç—á–∏–∫ —Å–ª–æ–≤, –Ω–∞–ø–∏—Å–∞–Ω–Ω—ã—Ö –∫–∞–ø—Å–æ–º
    for word in words:
        if word.isupper():  # –ü—Ä–æ–≤–µ—Ä–∫–∞, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ —Å–ª–æ–≤–æ –Ω–∞–ø–∏—Å–∞–Ω–Ω—ã–º –∫–∞–ø—Å–æ–º
            capitalized_count += 1
    return capitalized_count

# –ü—Ä–∏–º–µ–Ω–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –∫ —Å—Ç–æ–ª–±—Ü—É 'Text' –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ
df['—Åapitalized_word_num'] = df['text'].apply(count_capitalized_words)
```
