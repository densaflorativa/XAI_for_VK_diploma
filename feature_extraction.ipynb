```python
df.columns = ['text', 'post_date', 'user_id', 'emo', 'soc', 'psych', 'health',
       'emo_class', 'soc_class', 'psych_class', 'health_class','sentiment',
       'max_synt_tree_depth', 'min_synt_tree_depth', 'relative_clause_num', 'first_person_num','other_person_num','adjective_num', 'comparative_adjective_num',
'super_degree_num','indicative_mood_num','imperative_mood_num','imperfect_verbs_num','perfect_verbs_num', 'past_tense_verb_num',
'present_tense_verb_num','future_tense_verb_num', 'sent_words_num', 'mean_sent_words_num','max_sent_words_num','mean_words_len','max_words_num',
'epeated_words_num', 'most_repeated_word_num','emoji_num','pos_bracket_num', 'neg_bracket_num']
```

```python
import nltk

nltk.download('punkt')

def count_sentences(text):
    sentences = nltk.sent_tokenize(text)
    return len(sentences)

df['sent_num'] = df['text'].apply(count_sentences)
```

```python
import nltk
import string

# Функция для подсчета знаков препинания в предложении
def count_punctuation(sentence):
    count = 0
    for token in nltk.word_tokenize(sentence):
        if token in string.punctuation:
            count += 1
    punctuation_count = count / len(nltk.word_tokenize(sentence))
    return punctuation_count

df['punct_per_sent_num'] = df.text.apply(count_punctuation)
```

```python
import re
```

```python
# Функция для подсчета количества многоточий в тексте
def count_ellipsis(text):
    ellipsis_count = len(re.findall('\.{3,}', text))
    return ellipsis_count

# Применение функции к столбцу 'Text' в датафрейме
df['ellipsis_num'] = df['text'].apply(count_ellipsis)
```

```python
def count_capitalized_words(text):
    words = text.split()  # Разделение текста на отдельные слова
    capitalized_count = 0  # Счетчик слов, написанных капсом
    for word in words:
        if word.isupper():  # Проверка, является ли слово написанным капсом
            capitalized_count += 1
    return capitalized_count

# Применение функции к столбцу 'Text' в датафрейме
df['сapitalized_word_num'] = df['text'].apply(count_capitalized_words)
```
