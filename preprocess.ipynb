```python
import pandas as pd
```

```python
df = pd.read_csv('correct_tiny_features.csv')
df
```

```python
health = pd.read_csv('health.csv')
health
```

```python
health['user_id'] = health['ID']
```

```python
health['soc'] = health['—Å–æ—Ü–∏–∞–ª—å–Ω–æ–µ –±–ª–∞–≥–æ–ø–æ–ª—É—á–∏–µ mean']
health['emo'] = health['—ç–º–æ—Ü–∏–æ–Ω–∞–ª—å–Ω–æ–µ –±–ª–∞–≥–æ–ø–æ–ª—É—á–∏–µ mean']
health['psych'] = health['–ø—Å–∏—Ö–æ–ª–æ–≥–∏—á–µ—Å–∫–æ–µ –±–ª–∞–≥–æ–ø–æ–ª—É—á–∏–µ mean']
health['health'] = health['–ø—Å–∏—Ö–æ–ª–æ–≥–∏—á–µ—Å–∫–æ–µ –∑–¥–æ—Ä–æ–≤—å–µ mean']
```

```python
health2 = health[health.user_id.isin(df.user_id.unique())]
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# –£–∫–∞–∂–∏—Ç–µ –Ω–∞–∑–≤–∞–Ω–∏—è –∫–æ–ª–æ–Ω–æ–∫
columns = ['health', 'soc', 'emo', 'psych']

# –°–æ–∑–¥–∞–π—Ç–µ —Å–µ—Ç–∫—É –≥—Ä–∞—Ñ–∏–∫–æ–≤
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
axes = axes.flatten()

# –ü—Ä–æ—Ö–æ–¥ –ø–æ –∫–∞–∂–¥–æ–π –∫–æ–ª–æ–Ω–∫–µ –∏ —Å–æ–∑–¥–∞–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞
for i, column in enumerate(columns):
    # –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ
    sns.histplot(data=health2, x=column, kde=True, ax=axes[i])
    if column == 'health':
      axes[i].set_title(f'–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Å–∏—Ö–æ–ª–æ–≥–∏—á–µ—Å–∫–æ–µ –∑–¥–æ—Ä–æ–≤—å–µ')
    if column == 'soc':
      axes[i].set_title(f'–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å–æ—Ü–∏–∞–ª—å–Ω–æ–µ –±–ª–∞–≥–æ–ø–æ–ª—É—á–∏–µ')
    if column == 'emo':
      axes[i].set_title(f'–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —ç–º–æ—Ü–∏–æ–Ω–∞–ª—å–Ω–æ–µ –±–ª–∞–≥–æ–ø–æ–ª—É—á–∏–µ')
    if column == 'psych':
      axes[i].set_title(f'–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Å–∏—Ö–æ–ª–æ–≥–∏—á–µ—Å–∫–æ–µ –±–ª–∞–≥–æ–ø–æ–ª—É—á–∏–µ')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('–ß–∞—Å—Ç–æ—Ç–∞')

    # –°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
    mean_val = df[column].mean()
    std_val = df[column].std()
    axes[i].axvline(mean_val, color='red', linestyle='--', label='–°—Ä–µ–¥–Ω–µ–µ')
    axes[i].axvline(mean_val + std_val, color='green', linestyle='--', label='–°—Ä–µ–¥–Ω–µ–µ ¬± –°—Ç–¥ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ')
    axes[i].axvline(mean_val - std_val, color='green', linestyle='--')
    axes[i].legend()

# –û—Ç–æ–±—Ä–∞–∑–∏—Ç–µ –≥—Ä–∞—Ñ–∏–∫–∏
plt.tight_layout()
plt.show()
```
